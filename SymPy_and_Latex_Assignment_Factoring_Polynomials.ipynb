```python
from sympy import symbols, gcd, init_printing, pprint

# กำหนดตัวแปร
x = symbols('x')

# กำหนดพหุนาม
polynomial = x**3 - 6*x**8 + 10*x**10

# หาตัวประกอบร่วมมาก (GCD)
gcf = gcd(polynomial.args)

# แยกตัวประกอบ
result = gcf * (polynomial / gcf)

# แสดงผลลัพธ์
result
```
